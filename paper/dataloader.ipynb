```python
import pandas as pd

# List of CSV file paths
csv_files = ['../data/Results1.csv', '../data/Results2.csv', '../data/Results3.csv']

# Set to store unique column names
unique_column_names = set()

# Loop through each CSV file
for file in csv_files:
    # Load the CSV file
    df = pd.read_csv(file)
    
    # Get the column names and add them to the set
    unique_column_names.update(df.columns.dropna().str.replace(' ', '').tolist())

# Sort the unique column names chronologically
sorted_column_names = sorted(unique_column_names)

# Print the sorted column names
print("Unique Column Names (Chronologically):")
for column_name in sorted_column_names:
    print(column_name)

# Print the number of unique column names
print("Number of Unique Column Names:", len(unique_column_names))
```

```python
import torch
import csv
import numpy as np
import pandas as pd

class Data(torch.utils.data.Dataset):
    def __init__(self, label, features, csv_dir):
        self.features = features
        self.label = label
        content = self.read_csv(csv_dir)
        self.content = self.filter_incomplete_cases(content)
        self.x = np.array(self.content[self.features], dtype=np.float32)
        self.y = np.array(self.content[self.label], dtype=np.float32)

    def read_csv(self, csv_file):
        df = pd.read_csv(csv_file)
        df.columns = df.columns.str.replace(' ', '')  # Remove spaces from column names
        df['Factor1'] = df['Factor1'].map({'Factor1:CASE': 1, 'Factor1:CONTROL': 0})
        return df

    def filter_incomplete_cases(self, content):
        filtered_content = content.dropna(subset=[self.label])
        return filtered_content

    def __len__(self):
        return len(self.content)

    def __getitem__(self, idx):
        return self.x[idx], self.y[idx]

    def input_length(self):
        return len(self.__getitem__(0)[0])

    @property
    def all(self):
        return self.x, self.y

if __name__ == "__main__":
    csv_dir = '../data/Results1.csv'
    df = pd.read_csv(csv_dir)
    df.columns = df.columns.str.replace(' ', '')  # Remove spaces from column names
    
    # Check column names
    print("Column Names:")
    print(df.columns)
    
    # Update the label based on the desired column name
    label = 'MASS_RT'  # Modify this label if needed
    
    features = df.columns[1:-1]  # Exclude the first and last columns
    data = Data(label=label, features=features, csv_dir=csv_dir)

    # Print the loaded data
    print("Features:")
    print(data.x)
    print("Labels:")
    print(data.y)
```
